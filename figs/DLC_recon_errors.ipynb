```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sbj_d = {'a0f66459': [1.45, 4.27], 'c95c1e82': [1.44, 3.58], 'cb46fd46': [1.58, 6.95], 'fcb01f7a': [1.63, 6.02],
         'ffb52f92': [1.43, 3.42], 'b4ac1726': [1.43, 6.63], 'f3b79359': [1.51, 5.45], 'ec761078': [1.84, 10.35],
         'f0bbc9a9': [1.4, 4.05], 'abdb496b': [1.48, 7.59], 'ec168864': [1.51, 5.45], 'b45e3f7b': [1.52, 4.73]}

train_err = [val[0] for key, val in sbj_d.items()]
test_err = [val[1] for key, val in sbj_d.items()]

nsbjs = len(train_err)
sbj_nums = [val+1 for val in range(nsbjs)]
sbj = ['P'+str(val).zfill(2) for val in sbj_nums]
```

```python
fig = plt.figure(figsize=(3.75, 3), dpi=150)
plt.bar(sbj,train_err,color='dimgrey')
plt.bar(sbj,test_err,color='lightgrey')
plt.bar(sbj,train_err,color='dimgrey')
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().set_xticklabels(sbj, rotation=45)
plt.legend(['Train set','Holdout set'], frameon=False, fontsize=8)
plt.tick_params(labelsize=9)
plt.ylabel('Reconstruction error (pixels)')

fig_sp = '/data1/users/stepeter/mvmt_init/figs/'
fig_name = 'dlc_recon_errs'
dpi_plt = 300
fig.savefig(fig_sp+fig_name+'.png',
            format='png',transparent= True,dpi=dpi_plt,
            bbox_inches = 'tight', pad_inches = 0.01,
            )
plt.show()
```

```python
print(round(np.array(train_err).mean(),2))
print(round(np.array(train_err).std(),2))
print(round(np.array(test_err).mean(),2))
print(round(np.array(test_err).std(),2))
```
